Получить интуицию, на что влияют параметры распределений. Построить по 5 функций плотностей с разными параметрами для каждого распределения: нормальное, экспоненциальное, Стьюдента на одном графике.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts

%matplotlib inline
```

```python
# нормальное распределение
# loc - mu, scale - среднеквадратичное отклонение
x = np.linspace(-2,4,100)
norm_rv_1 = sts.norm(loc=0, scale=0.9)
norm_rv_2 = sts.norm(loc=0, scale=0.5)
norm_rv_3 = sts.norm(loc=1, scale=2)
norm_rv_4 = sts.norm(loc=1, scale=1)
norm_rv_5 = sts.norm(loc=3, scale=0.3)

pdf1 = norm_rv_1.pdf(x)
pdf2 = norm_rv_2.pdf(x)
pdf3 = norm_rv_3.pdf(x)
pdf4 = norm_rv_4.pdf(x)
pdf5 = norm_rv_5.pdf(x)
```

```python
plt.figure(figsize=(15,10))

plt.plot(x, pdf1, label="pdf1")
plt.plot(x, pdf2, label="pdf2")
plt.plot(x, pdf3, label="pdf3")
plt.plot(x, pdf4, label="pdf4")
plt.plot(x, pdf5, label="pdf5")

plt.ylabel('$f(x)$')
plt.xlabel('$x$')
plt.legend()


plt.show()
```

Наблюдения: 

Параметр loc $\mu$ - смещает центр графика по оси Х

Парметр среднеквадратичное отклонение $\sigma$ - влияет на разброс значений относительно центра, чем больше $\sigma$ тем график ниже относительно оси Y и шире отсноительно оси Х

```python
# распределение экспоненциальное
# lam - смещение

x = np.linspace(-2,4,100)
expon_rv1 = sts.expon(0.1)  #lam
expon_rv2 = sts.expon(0.3)
expon_rv3 = sts.expon(-0.5)
expon_rv4 = sts.expon(1)
expon_rv5 = sts.expon(-1.5)

ex_pdf1 = expon_rv1.pdf(x)
ex_pdf2 = expon_rv2.pdf(x)
ex_pdf3 = expon_rv3.pdf(x)
ex_pdf4 = expon_rv4.pdf(x)
ex_pdf5 = expon_rv5.pdf(x)
```

```python
plt.figure(figsize=(15,10))

plt.plot(x, ex_pdf1, label="ex_pdf1")
plt.plot(x, ex_pdf2, label="ex_pdf2")
plt.plot(x, ex_pdf3, label="ex_pdf3")
plt.plot(x, ex_pdf4, label="ex_pdf4")
plt.plot(x, ex_pdf5, label="ex_pdf5")

plt.ylabel('$f(x)$')
plt.xlabel('$x$')
plt.legend()


plt.show()
```

Наблюдения: 

Параметр $\lambda$ - влияиет на смещение графика относительно оси X

```python
#  распределение Стьюдента

x = np.linspace(-3,3,100)

t_rv1 = sts.t(1)
t_rv2 = sts.t(5)
t_rv3 = sts.t(10)
t_rv4 = sts.t(30)
t_rv5 = sts.t(1000)

t_pdf1 = t_rv1.pdf(x)
t_pdf2 = t_rv2.pdf(x)
t_pdf3 = t_rv3.pdf(x)
t_pdf4 = t_rv4.pdf(x)
t_pdf5 = t_rv5.pdf(x)
```

```python
plt.figure(figsize=(15,10))

plt.plot(x, t_pdf1, label="t_pdf1")
plt.plot(x, t_pdf2, label="t_pdf2")
plt.plot(x, t_pdf3, label="t_pdf3")
plt.plot(x, t_pdf4, label="t_pdf4")
plt.plot(x, t_pdf5, label="t_pdf5")


plt.ylabel('$f(x)$')
plt.xlabel('$x$')
plt.legend()


plt.show()
```

Наблюдения: 

График напоминает форму колокола как у стандартного нормального распределения, но немного ниже и шире. 
Чем больше степеней свободы - тем кривая функции все больше напоминает стандартное нормальное распределение.
То есть чем больше N тем выше и уже график кривой.
